```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
df = pd.read_csv('prices_mv.csv')
df
```

```python
df.bedrooms.fillna(math.floor(df.bedrooms.median()), inplace=True)
x1_mean = df.area.mean()
x2_mean = df.bedrooms.mean()
x3_mean = df.age.mean()
y_mean = df.price.mean()
```

```python
x1_mean, x2_mean, x3_mean, y_mean
```

```python
df
```

```python
df['x1-X1'] = df['area']-x1_mean
df['x2-X2'] = df['bedrooms']-x2_mean
df['x3-X3'] = df['age']-x3_mean
df
```

```python
df['x1-X1_sq'] = df['x1-X1']*df['x1-X1']
df['x2-X2_sq'] = df['x2-X2']*df['x2-X2']
df['x3-X3_sq'] = df['x3-X3']*df['x3-X3']
df
```

```python
df['y-Y'] = df['price']-y_mean
df
```

```python
m1 = sum(df['x1-X1']*df['y-Y'])/sum(df['x1-X1_sq'])
m2 = sum(df['x2-X2']*df['y-Y'])/sum(df['x2-X2_sq'])
m3 = sum(df['x3-X3']*df['y-Y'])/sum(df['x3-X3_sq'])
m1, m2, m3
```

```python
#y_mean = x1_mean*m1 + x2_mean*m2 + x3_mean*m3
c = y_mean-(x1_mean*m1)-(x2_mean*m2)-(x3_mean*m3)
c
```
