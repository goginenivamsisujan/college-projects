```python
import numpy as np
import pandas as pd
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x, y):
    m = c = 0
    iterations = 10000
    n = len(x)
    learningRate = 0.01
    for i in range(iterations):
        y_pred = m*x+c
        cost = 1/n*sum(val ** 2 for val in (y-y_pred))
        md = -2*sum(x*(y-y_pred))
        cd = -2*sum(y-y_pred)
        m = m - learningRate * md
        c = c - learningRate * cd
        print("m : {}, c : {}, cost : {}".format(m, c, cost))
```

```python
gradient_descent(x,y)
```
